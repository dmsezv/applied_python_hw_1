```python
import pandas as pd
import numpy as np

# Реальные средние температуры (примерные данные) для городов по сезонам
seasonal_temperatures = {
    "New York": {"winter": 0, "spring": 10, "summer": 25, "autumn": 15},
    "London": {"winter": 5, "spring": 11, "summer": 18, "autumn": 12},
    "Paris": {"winter": 4, "spring": 12, "summer": 20, "autumn": 13},
    "Tokyo": {"winter": 6, "spring": 15, "summer": 27, "autumn": 18},
    "Moscow": {"winter": -10, "spring": 5, "summer": 18, "autumn": 8},
    "Sydney": {"winter": 12, "spring": 18, "summer": 25, "autumn": 20},
    "Berlin": {"winter": 0, "spring": 10, "summer": 20, "autumn": 11},
    "Beijing": {"winter": -2, "spring": 13, "summer": 27, "autumn": 16},
    "Rio de Janeiro": {"winter": 20, "spring": 25, "summer": 30, "autumn": 25},
    "Dubai": {"winter": 20, "spring": 30, "summer": 40, "autumn": 30},
    "Los Angeles": {"winter": 15, "spring": 18, "summer": 25, "autumn": 20},
    "Singapore": {"winter": 27, "spring": 28, "summer": 28, "autumn": 27},
    "Mumbai": {"winter": 25, "spring": 30, "summer": 35, "autumn": 30},
    "Cairo": {"winter": 15, "spring": 25, "summer": 35, "autumn": 25},
    "Mexico City": {"winter": 12, "spring": 18, "summer": 20, "autumn": 15},
}

# Сопоставление месяцев с сезонами
month_to_season = {12: "winter", 1: "winter", 2: "winter",
                   3: "spring", 4: "spring", 5: "spring",
                   6: "summer", 7: "summer", 8: "summer",
                   9: "autumn", 10: "autumn", 11: "autumn"}

# Генерация данных о температуре
def generate_realistic_temperature_data(cities, num_years=10):
    dates = pd.date_range(start="2010-01-01", periods=365 * num_years, freq="D")
    data = []

    for city in cities:
        for date in dates:
            season = month_to_season[date.month]
            mean_temp = seasonal_temperatures[city][season]
            # Добавляем случайное отклонение
            temperature = np.random.normal(loc=mean_temp, scale=5)
            data.append({"city": city, "timestamp": date, "temperature": temperature})

    df = pd.DataFrame(data)
    df['season'] = df['timestamp'].dt.month.map(lambda x: month_to_season[x])
    return df

# Генерация данных
data = generate_realistic_temperature_data(list(seasonal_temperatures.keys()))
data.to_csv('temperature_data.csv', index=False)
```

```python
import dask.dataframe as dd
import time

df = pd.read_csv("temperature_data.csv")
df["timestamp"] = pd.to_datetime(df["timestamp"])
df = df.sort_values(by=["city", "timestamp"])

ndf = df.copy()
ddf = dd.from_pandas(df, npartitions=4)

def calculate_rolling_features(group):
    group["rolling_mean"] = group["temperature"].rolling(window=30, min_periods=1).mean()
    group["std_dev"] = group["temperature"].rolling(window=30, min_periods=1).std()
    std_dev_2 = 2 * group["std_dev"]
    group["anomaly"] = (group["temperature"] < (group["rolling_mean"] - std_dev_2)) | (group["temperature"] > (group["rolling_mean"] + std_dev_2))
    return group


start_time = time.time()
ndf = ndf.groupby("city").apply(calculate_rolling_features, include_groups=False)
end_time = time.time()
print(f"No parallel: {end_time - start_time:.2f} s")

meta = ndf.head(0)

start_time = time.time()
ddf = ddf.groupby("city").apply(calculate_rolling_features, meta=meta, include_groups=False)
result_df = ddf.compute()
end_time = time.time()
print(f"Parallel: {end_time - start_time:.2f} s")

result_df.to_csv("temperature_analys.csv")
```

```python
import aiohttp
import asyncio
import nest_asyncio

API_KEY = ""
BASE_URL = "http://api.openweathermap.org"

CITIES = ["Moscow", "Berlin", "Paris", "London", "New York"]

month_to_season = {
        12: "winter", 
        1: "winter", 
        2: "winter",

        3: "spring", 
        4: "spring", 
        5: "spring",

        6: "summer", 
        7: "summer", 
        8: "summer",
        
        9: "autumn", 
        10: "autumn", 
        11: "autumn"
    }

current_season = month_to_season[pd.Timestamp.now().month]


def process_city(city, current_temp):
    if current_temp is None:
        print(f"Current temperature in {city} is None")
        return

    df = pd.read_csv("temperature_data.csv")
    df["timestamp"] = pd.to_datetime(df["timestamp"])
    df = df.sort_values(by=["city", "timestamp"])

    historical_data = df[(df["city"] == city) & (df["season"] == current_season)]
    mean_temp = historical_data["temperature"].mean()
    std_dev = historical_data["temperature"].std()

    if current_temp < (mean_temp - 2 * std_dev) or current_temp > (mean_temp + 2 * std_dev):
        print("Anomaly for this season")
    else:
        print("Normal for this season")

    print(f"Mean historical temperature: {mean_temp:.2f} °C\n")




def get_current_temperature_sync(city):
    url = f"{BASE_URL}/data/2.5/weather?q={city}&appid={API_KEY}&units=metric"
    response = requests.get(url)
    data = response.json()
    if response.status_code == 200:
        return data["main"]["temp"]
    else:
        print(f"[ERROR]: {data['message']}")
        return None


async def get_current_temperature_async(session, city):
    url = f"{BASE_URL}/data/2.5/weather?q={city}&appid={API_KEY}&units=metric"
    async with session.get(url) as response:
        data = await response.json()
        if response.status == 200:
            return data["main"]["temp"]
        else:
            print(f"[ERROR]: {data['message']}")
            return None


def check_anomalies_sync():
    for city in CITIES:
        current_temp = get_current_temperature_sync(city)
        process_city(city, current_temp)


async def check_anomalies_async():
    async with aiohttp.ClientSession() as session:
        tasks = [get_current_temperature_async(session, city) for city in CITIES]
        results = await asyncio.gather(*tasks)

        for city, current_temp in zip(CITIES, results):
            process_city(city, current_temp)


start_time = time.time()
check_anomalies_sync()
end_time = time.time()
print(f"Sync: {end_time - start_time:.2f} s")
print("\n--------------------------------\n")
start_time = time.time()
asyncio.run(check_anomalies_async())
end_time = time.time()
print(f"Async: {end_time - start_time:.2f} s")
```

Вывод: Асинхронный подход предпочтителен для задач, требующих выполнения большого количества сетевых запросов, так как он позволяет значительно сократить время выполнения.
